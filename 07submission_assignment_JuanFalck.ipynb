## Some interesting things with the dataset - Principal Component Analysis

```python
#We will do something interesting with the data and do ML.  We will reduce dimensions using PCA (Principal
# Component Analysis) this will allow us to plot using two dimension and see if we see some common
# groupings on the different wine classes based on their respective
# attributes
```

```python
df3.head()
```

```python
#We will separate the data into two tables. One only with the row's wine class and
# another table will all the numeric data
df3_data = df3.iloc[:,1:14]
df3_classes = df3["wine_class"]


d = {'A':1, 'B':2,"C":3}
df3_classes= df3_classes.map(d)
df3_classes.head()
```

```python
# We will scikit-learn to do PCA.
#First will standardize the data to perform PCA.

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df3_data)
scaled_data = scaler.transform(df3_data)
```

```python
# Here we tell sklearn to reduce the datasset to 2 Principal components
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
#Let's plot the data, now reduced to only two dimensions and colore it based on its wine class
#and see if we see any common grouping (Clustering)
from matplotlib import pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=df3_classes,cmap='rainbow')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
# We can clearly see that the 3 wine classes are nicely clustered in three distinct groups
# based on its reduced dimension (Principal Components)
```

```python
import seaborn as sns
```

```python
#Here we will plot the Principal components and its
#corelation to each Column in the original dataset

col_names =  ["alcohol","malic_acid","ash","alcalinity","magnesium","phenols","flavanoids",
               "non_flavanoids","proanthocyanins","color_instensity","hue","Protein","proline"]
map= pd.DataFrame(pca.components_,columns=col_names)
plt.figure(figsize=(12,6))
sns.heatmap(map,cmap='bwr')
```
