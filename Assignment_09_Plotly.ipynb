# Data 602 - Assignment 9 - Plotly

### Import packages and load tips data

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "svg"

df = px.data.tips()
df.head()
```

## Bar Plot

```python
fig = px.bar(df, x="sex", y="total_bill",
             color='smoker', barmode='group',
             height=400)
fig.show()
```

```python
fig = px.bar(df, x="sex", y="total_bill",
             color='day', barmode='group',
             height=400)
fig.show()
```

## Scatter Plot

```python
fig = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker")
fig.show()
```

```python
fig = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker",trendline="ols")
fig.show()
```

```python
fig = px.scatter(df, x="total_bill", y="tip", facet_col="day", facet_row="time")
fig.show()
```

## Histogram

```python
fig = px.histogram(df, x="tip", marginal="rug", # can be `box`, `violin`
                         hover_data=df.columns)
fig.show()
```

## Boxplot

```python
fig = px.box(df, x="smoker", y="tip", color="smoker")
fig.show()
```
